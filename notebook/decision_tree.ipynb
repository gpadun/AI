## CART algorithm
We're going to use a Classification and Regression Tress algorithm. It basically works as adding a node to the tree, putting all the training data through this node, asking a boolean question, that means, a kind of yes or no question, then we split the data into subsets, that will be the inputs for the following nodes. Our goal is to find the best questions at each node, the questions that best unmix the data. We can do that mesuring how much a question really does that well.


## The Gini Impurity
It measures the probability to incorrectly label an data, randomly, from the dataframe. If the data are all from the same label, the impurity will be 0.
Its formula is 1 - Sum of probabilities of all labels squared.


```python
def gini(df):
  impurity = 1
  prob_sum = 0
  data_lableds = df['species'].value_counts().to_list()
  probabilities = []
  for value in data_lableds:
    probabilities.append(value/float(df.shape[0]))
  for probability in probabilities:
    probability = probability**2
    prob_sum += probability
  impurity -= prob_sum
  return impurity
```

## Information gain
It is basically a number that tells how much a node helps to unmix the labels at a node. It goes as follows:
- First we calculate the impurity of our dataset
- Then for each node, we calculate the impurity of the child nodes, and take a weighted average of that result;
- Finally, we subtract this with the initial impurity;     

That is how we get the information gain.
We will use it to select the best questions at each point of the tree.

The formula is IG = Gini_parent​ − (left_w​eight × Gini_left​ + right_weight ​× Gin_right​)



```python
def information_gain(left_node, right_node, current_impurity):
  weight_left = float(len(left_node)) / (float(len(left_node)) + float(len(right_node)))
  weight_right = 1 - weight_left
  return current_impurity - (weight_left*gini(left_node) + weight_right*gini(right_node))
```
